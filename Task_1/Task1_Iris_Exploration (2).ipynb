```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set plotting defaults
plt.rcParams['figure.dpi'] = 120
sns.set(style='whitegrid')
```

## 1. Dataset Loading

We will use Seaborn's built-in Iris dataset. The dataset contains 150 samples with 4 numeric features and a categorical species label.


```python
iris_df = sns.load_dataset('iris')

print('Shape:', iris_df.shape)
print('Columns:', iris_df.columns.tolist())

# Show first 5 rows
iris_df.head()
```

## 2. Dataset Inspection

Check data types, missing values, and summary statistics.

```python
# Data types and missing values
iris_df.info()

# Summary statistics
iris_df.describe()

# Missing values
print('\nMissing values per column:\n', iris_df.isnull().sum())
```

## 3. Data Visualization

### 3.1 Scatter Plot — Sepal Length vs Petal Length

Observation: Visual separation of species; especially *setosa*.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species', s=80)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

### 3.2 Histograms — Feature Distributions

```python
iris_df.hist(figsize=(10,8), bins=15)
plt.suptitle('Feature Distributions')
plt.show()
```

### 3.3 Box Plots — Outlier Detection

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris_df.drop(columns='species'))
plt.title('Box Plots for Numerical Features')
plt.show()
```

### 3.4 Pair Plot — Pairwise Relationships

Pairplots show relationships between every pair of features, colored by species.

```python
sns.pairplot(iris_df, hue='species', diag_kind='hist')
plt.suptitle('Pairwise Relationships (Pairplot)', y=1.02)
plt.show()
```

### 3.5 Correlation Heatmap — Feature Correlations

```python
plt.figure(figsize=(6,5))
corr = iris_df.drop(columns='species').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='Blues')
plt.title('Feature Correlation Matrix')
plt.show()

corr
```

## 4. Insights & Observations

- The dataset has **150 rows** and **5 columns** (4 numeric, 1 categorical).
- There are **no missing values**.
- **Petal length** and **petal width** have the strongest correlation and best separate species.
- **Setosa** is linearly separable from the other two species using petal features.
- Mild outliers are visible in `sepal_width`, but they are not problematic for basic models.
